```python
import pycountry
```

```python
def do_fuzzy_search(country):
    try:
        result = pycountry.countries.search_fuzzy(country)
    except Exception:
        return np.nan
    else:
        return result[0].alpha_3
```

```python
iso_map = {country: do_fuzzy_search(country) for country in happy_transformed["Country"].unique()}

happy_transformed["code"] = happy_transformed["Country"].map(iso_map)
```

```python
happy_transformed.dropna(subset = ["code"], inplace=True)
```
